```python
from selenium import webdriver
import time
import pandas as pd
import numpy as np
```

```python
driver = webdriver.Chrome(
        executable_path=("C:/Users/רוני/PycharmProjects/BONOSHAGOREM/chromedriver.exe"))
driver.get("https://www.wisebuy.co.il/products.aspx?category=c-tabletpc")

dict_info = {}
#get information
all_tablets = driver.find_elements_by_class_name("ProdGeneralInfo")
total = len(all_tablets)
num_of_tablets = total

for tablet in all_tablets:
    if num_of_tablets == 0:
        break
    
    pairs = tablet.find_elements_by_class_name("pair")
    for info in pairs:
        category, value = info.text.split("\n")
        if category not in dict_info.keys():
            dict_info[category] = [None]*total
        dict_info[category][total - num_of_tablets] = value
    num_of_tablets -= 1

#get grades
num_of_tablets = total
dict_info['grades'] = [None]*num_of_tablets
stars_list = driver.find_elements_by_class_name("Reviews")
for stars_info in stars_list:
    if num_of_tablets == 0:
        break
    g = stars_info.find_elements_by_class_name("Stars")
    for i in g:
        temp = i.get_attribute("title").split('ציון גולשים')[1]
        if temp != ' ':
            dict_info['grades'][total - num_of_tablets] = int(temp)
            
    num_of_tablets -= 1


#get names
num_of_tablets = total
dict_info['names'] = [None]*num_of_tablets
names_list = driver.find_elements_by_class_name("ProdInfoTitle")
for name in names_list:
    if num_of_tablets == 0:
        break
    title = name.text
    dict_info['names'][total-num_of_tablets] = title
    num_of_tablets -= 1


#get price
num_of_tablets = total
dict_info['prices'] = [None]*num_of_tablets
price_list = driver.find_elements_by_class_name("pricesTxt")
for name in price_list:
    if num_of_tablets == 0:
        break
    price = name.text
    dict_info['prices'][total-num_of_tablets] = price
    num_of_tablets -= 1
data = pd.DataFrame(dict_info)
```

```python
cleaned_data = data.drop(['מהירות מעבד:','מצלמה קדמית:', 'מצלמה ראשית:','סוג המעבד:','מספר ליבות:','רזולציית מסך:','גרסת מערכת הפעלה:','תמיכה ברשת סלולרית:',], axis=1)
cleaned_data['grades'] = cleaned_data['grades'].fillna(int(-100))
finel = cleaned_data[cleaned_data['grades']>=0]
finel.head(15)
```

```python
def df_column_switch(df, column1, column2):
    i = list(df.columns)
    a, b = i.index(column1), i.index(column2)
    i[b], i[a] = i[a], i[b]
    df = df[i]
    return df
```

```python
for_query = finel.drop(['grades'], axis=1)
for_query = df_column_switch(for_query,"names","prices")
for_query.head(15)
```
